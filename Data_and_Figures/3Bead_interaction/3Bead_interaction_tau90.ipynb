```python
%load_ext autoreload
%autoreload 2

import sys
sys.path.append("..")
import pandas
import numpy as np
import matplotlib.pyplot as plt
import os.path
import numbers

datadir = '/Users/billiemeadowcroft/Dropbox/PhDGithub/RemoteTether/Data_and_Figures/3Bead_interaction/Data/'
plotsdir = '/Users/billiemeadowcroft/Dropbox/PhDGithub/RemoteTether/Data_and_Figures/3Bead_interaction/Plots/'
clrs = ['#2c7fb8','#41b6c4','#7fcdbb']
```

```python
clr2 = 'mediumseagreen'
clr1 = 'mediumseagreen' #cmap((8)/(12))
clr4 = 'tab:blue' #cmap2((10)/(12))
clr3 = 'tab:blue'
```

```python
fname0 = 'exp_r30Tau90.txt'
fname1 = 'exp_r18Tau90.txt'
fname2 = 'sim_r3.4Tau90.txt'
fname3 = 'sim_r1.8Tau90.txt'
intexp30 = pandas.read_csv(datadir  + fname0, sep=' ', header = 'infer')
ellN  = intexp30.ell
interaction30N =  intexp30.excess

intexp30 = pandas.read_csv(datadir  + fname1, sep=' ', header = 'infer')
ell  = intexp30.ell
interaction18 =  intexp30.excess

intsim3 = pandas.read_csv(datadir  + fname2, sep=' ', header = 'infer')
ellsim3  = intsim3.ell
sim3int =  intsim3.excessr
sim3int_std = intsim3.exc_std

intsim1 = pandas.read_csv(datadir  + fname3, sep=' ', header = 'infer')
ellsim1  = intsim1.ell
sim1int =  intsim1.excessr
sim1int_std = intsim1.exc_std
```

```python
fig,ax= plt.subplots(1,figsize = (4,4))
#fig2,ax2= plt.subplots(1,figsize = (5,4))
ax.plot(ellN,interaction30N, color = clr4,linewidth  = 3,marker = 'X',markersize = 5,linestyle = ':',label = 'Experiment $r = 3.0~ D$')
ax.plot(ell,interaction18,color = clr1,linewidth  = 3,marker = 'o',markersize = 5,linestyle = ':',label = 'Experiment $r = 1.8~ D$')
ax.plot(ellsim3[1:],sim3int[1:],color = clr4,linewidth  = 2,marker = 'X',linestyle = '-',markersize = 7, label = "Simulation $r = 3.4~ D$")
ax.fill_between(ellsim3[1:],np.add(sim3int[1:],-sim3int_std[1:]),np.add(sim3int[1:],sim3int_std[1:]), color = clr4,alpha = 0.2)
ax.plot(ellsim1[1:],sim1int[1:],color = clr1,linewidth  = 2,marker = 'X',linestyle = '-',markersize = 7, label = "Simulation $r = 1.6~ D$")
ax.fill_between(ellsim1[1:],np.add(sim1int[1:],-sim1int_std[1:]),np.add(sim1int[1:],sim1int_std[1:]), color = clr1,alpha = 0.2)
ax.set_xlabel(r'$\ell~ [D]$', fontsize  = 14)
ax.legend(frameon=False, loc = [0.02,0.69])
ax.set_ylabel(r'$\Delta E$ [$k_{\mathrm{B}}T$]', fontsize  = 14)
ax.set_ylim([-6,6.5])
ax.set_xlim([2.,6.])
fig.tight_layout()
plotname = '3beadInteraction_tau90.pdf'
fig.savefig(plotsdir+plotname)
```
